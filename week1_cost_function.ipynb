```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Model Representation
x_train = np.array([1.0, 2.0])
y_train = np.array([300, 500])
print(f"x_train =  {x_train}")
print(f"y_train =  {y_train}")
```

```python
print(f"x_train.shape =  {x_train.shape[0]}")
print(f"y_train.shape =  {y_train.shape[0]}")
m = x_train.shape[0]
print(f"Number of training examples: {m}")
```

```python
i = 0 #iterating through the trianing set
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^{i}, y^{i}) = (x^{x_i}, y^{y_i})")
```

```python
#plotting time
plt.scatter(x_train, y_train, marker = 'x', c='r')
plt.title("Housing Prices")
plt.ylabel("Price in 1000s of dollars")
plt.xlabel("size in 1000 square feet")
plt.show()
```

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b): # the equation is f_wb = w * x[0] + b
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label="Actual Values")
plt.title("Housing Prices")
plt.ylabel("Price in 1000s of dollars")
plt.xlabel("size in 1000 square feet")
plt.legend()
plt.show()
```
